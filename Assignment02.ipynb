<h3>1. Function Definition</h1>

$\large{f(x) = x^3}$

```python
import numpy as np
from matplotlib import pyplot as plt  


def f(x):
    return x*x*x

x = [a for a in range (-19,20)]
y = [f(b) for b in range(-19,20)]

plt.plot(x,y,'b',label = 'f(x) = x^3')

plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Taylor Approximation')

plt.grid()
plt.legend()


plt.show()
```

<h3>2. Select a point</h3>

$\large{a = 10}$

```python
a = 10
plt.plot(x,y,'b',label = 'f(x) = x^3')
plt.plot(a,f(a),'ro',label = 'a = (10,1000)')

plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Taylor Approximation')

plt.grid()
plt.legend()

plt.show()
```

<h3>3. Define first-order Taylor approximation at the selected point.</h3>

$\large{n=1,\,a=10\,\,Taylor\,Approximation}$


$\large{\hat{f}(x) = 300x - 2000}$

```python
#n=1 a=10  Taylor Approximation   
def Taylor(a,h): # a : Taylor Approximate  , h : x
    x = [c for c in range(-19,20)]
    yhat = [300 * x - 2000 for x in range(-19,20)]
    plt.plot(x,yhat,'r',label = 'f(x) hat')

Taylor(a,1)
plt.plot(x,y,'b',label = 'f(x) = x^3')
plt.plot(a,f(a),'ro',label = 'a = (10,1000)')

plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Taylor Approximation')

plt.grid()
plt.legend()

plt.show()
```

$\large{n=2,\,a=10\,\,Taylor\,Approximation}$

$\large{\hat{f}(x) = 30x^2 - 300x + 1000}$

```python
#n=2 a=10  Taylor Approximation   
def Taylor2(a,h): # a : Taylor Approximate  , h : x
    x = [c for c in range(-19,20)]
    yhat = [30*x*x - 300*x + 1000 for x in range(-19,20)]
    plt.plot(x,yhat,'r',label = 'f(x) hat')

Taylor2(a,1)
plt.plot(x,y,'b',label = 'f(x) = x^3')
plt.plot(a,f(a),'ro',label = 'a = (10,1000)')

plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Taylor Approximation')

plt.grid()
plt.legend()

plt.show()
```
