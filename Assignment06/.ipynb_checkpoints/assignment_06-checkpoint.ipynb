```python
import PIL.Image as pilimg
import matplotlib.pyplot as plt
import numpy as np
import random
 
size_row = 50
size_col = 80

horizontal = np.zeros((size_row,size_col),dtype = 'int')
vertical = np.zeros((size_row,size_col),dtype = 'int')

for i in range(size_row):
    for j in range(size_col):
        horizontal[i][j] = i
        vertical[i][j] = j
        
#norm 1
def Mehnert_norm(x,y):
    d = (x - y)
    s = np.sum(np.abs(d))
    return(s)

#norm 2
def Euclidean_norm(x, y):
    d = (x - y) ** 2
    s = np.sum(d)
    r = np.sqrt(s)
    return(r)

def initial(centeroids,k):
    for i in range(k):
        for j in range(2):
            centeroids[i,j] = random.randrange(0,size_row)

            
table_label  = np.zeros((size_row , size_col), dtype='int')
#find min distance
def min_dst(temp,k,x,y):
    min = temp[0]
    table_label[x,y] = 0
    for i in range(k):
        if temp[i] < min:
            min = temp[i]
            table_label[x,y] = i  #centroid of x-th data is i
    return min

def Energy(energy):
    plt.title('Energy curve for each $K$')
    plt.plot(energy,'ro-')
    plt.show()
```

```python
# plot Energy
def visualize():
    plt.title('K-Means')
    imshow(table_label, cmap='BrBG')
    
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend() #범례
    plt.show()
```

```python
def k_means(norm,k):
    temp = np.zeros(k,dtype = 'float')
    
    centeroids = np.zeros((k,2),dtype = 'int')
    pre_centeroids = np.zeros((k,2),dtype = 'int')
    energy = []
    
    iteration = 0
    dst = np.zeros((size_row,size_col),dtype = 'float')
    
    initial(centeroids,k)
    
    while(not(np.array_equal(centeroids,pre_centeroids))):
        avg = np.zeros((k,2),dtype='int')
        cnt = np.zeros(k,dtype='int')   # avg / cnt
        
        pre_centeroids = centeroids
        iteration += 1
        
        #calculate each pixel and centeroid ' s distance(group) 
        for i in range(size_row):
            for j in range(size_col):
                for m in range(k):
                    xy = np.array((horizontal[i][j],vertical[i][j])) #hor+ver
                    if norm == "Mehnert":
                        temp[m] = Mehnert_norm(xy,centeroids[m,:])
                    elif norm == "Euclidean":
                        temp[m] = Euclidean_norm(xy,centeroids[m,:])
                    else:
                        return
                dst[i,j] = min_dst(temp,k,i,j)
            
        #change centeroids
        for i in range(size_row):
            for j in range(size_col):
                avg[table_label[i,j],0] += horizontal[i][0]
                avg[table_label[i,j],1] += vertical[0][j]
                cnt[table_label[i,j]] += 1
                
        for i in range(k):
            if cnt[i] != 0:
                avg[i,:] = avg[i,:]/cnt[i]
        
        centeroids = avg
        energy.append(np.sum(dst))
    
    visualize()
    Energy(energy)
    print(iteration)
```

```python
#norm1 , k = 3
k_means("Mehnert",3)
```

```python
#norm1 , k = 10
k_means("Mehnert",10)
```

```python
#norm1 , k = 3
k_means("Mehnert",15)
```

```python
#norm2 , k = 3
k_means("Euclidean",3)
```

```python
#norm2 , k = 10
k_means("Euclidean",10)
```

```python
#norm2 , k = 15
k_means("Euclidean",15)
```
